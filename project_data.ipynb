```python
import pandas as pd
import matplotlib.pyplot as plt

def bar_chart(col):
    Approved = train_data[train_data["Loan_Status"] == "Y"][col].value_counts()
    Disapproved = train_data[train_data["Loan_Status"] == "N"][col].value_counts()

    # Combining Approved and Disapproved into a DataFrame
    df1 = pd.DataFrame({'Approved': Approved, 'Disapproved': Disapproved})

    # Handling missing values (if any)
    df1 = df1.fillna(0)

    # Plotting the bar chart
    df1.plot(kind="bar", figsize=(10, 6))
    plt.title(f'Loan Approval Status by {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.legend(title='Loan Status')
    plt.show()

# Example usage:
# bar_chart('Credit_History')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
# encoded the gender column of our dataframe
# 1 -> Male, 0 ->  Female
encoder.fit(train_data['Gender'])
train_data['Gender'] = encoder.transform(train_data['Gender'])
```

```python
# encoded the married column of our dataframe
# 1 -> Yes, 0 -> No
encoder.fit(train_data['Married'])
train_data['Married'] = encoder.transform(train_data['Married'])
```

```python
# encoded the education column of our dataframe
# 0 -> Graduate, 1 -> Not Graduate
encoder.fit(train_data['Education'])
train_data['Education'] = encoder.transform(train_data['Education'])
```

```python
# encoded the self_employed column of our dataframe
# 1 -> Yes, 0 -> No
encoder.fit(train_data['Self_Employed'])
train_data['Self_Employed'] = encoder.transform(train_data['Self_Employed'])
```

```python
# encoded the property_area column of our dataframe
# 2 -> Urban, 0 -> Rural, 1 -> Semiurban 
encoder.fit(train_data['Property_Area'])
train_data['Property_Area'] = encoder.transform(train_data['Property_Area'])
```

```python
# encoded the loan_status column of our dataframe
# 1 -> Y, 0 -> N
encoder.fit(train_data['Loan_Status'])
train_data['Loan_Status'] = encoder.transform(train_data['Loan_Status'])
```

#### **Feature Selection**

```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
# Independent Features
X = train_data.iloc[:, :10]

# Dependent Features
y = train_data.iloc[:, 10]
```

```python
X.head(5)
```

```python
y
```

```python
best_features = SelectKBest(score_func = chi2, k = 5)
fit = best_features.fit(X, y)
```

```python
scores = pd.DataFrame(fit.scores_)
columns = pd.DataFrame(X.columns)
```

```python
feature_scores = pd.concat([scores, columns], axis = 1)
feature_scores.columns = ['Scores', 'features'] 
```

```python
feature_scores
```

```python
features_useful = feature_scores.nlargest(5, 'Scores')
```

```python
features_useful
```

```python
X = train_data[features_useful['features']]
```

```python
X.head(5)
```

#### **Scaling the data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

#### **Train Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

#### **Implementing the model using SVM**

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel = 'linear')
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

#### **Performance of Model**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_svc = accuracy_score(y_test, y_pred)
confusion_matrix_svc = confusion_matrix(y_test, y_pred)
```

```python
cr_svc = classification_report(y_test, y_pred)
```

```python
print(f"Accuracy Score of SVC :\n{accuracy_svc}\n")
print(f"Confusion Matrix of SVC :\n{confusion_matrix_svc}\n")
print(f"Classification Report of SVC :\n{cr_svc}")
```

#### **Cross Validation of SVC**

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_scores = cross_val_score(svc, X, y, cv = 10, scoring = 'accuracy')
```

```python
np.mean(cv_scores)
```
